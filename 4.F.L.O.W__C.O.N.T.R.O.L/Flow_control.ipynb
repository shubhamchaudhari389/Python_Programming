## 1. Conditional Statement:

### 1) If

if condition : statement<br>

or<br>

if condition:<br>
     Statement 1<br>
     Statement 2<br>
     Statement 3<br>
If Condition is True Then Statement Will Be Executed<br>


```python
name = input("Enter Name:")
if name == "Shubham":
    print("Hello Shubham Good Morning")
print("How Are You!!!")
```

### 2) if-else:
if condition :<br>
    Action-1<br>

else :<br>
    Action-2<br>

If condition is True Then Action-1 Will Be Executed Otherwise Action-2 Will Be Executed.<br>

```python
name = input("Enter The Name:")
if name == "Shubham":
    print("Hello Shubham Good Moring")
else:
    print("Hello Guest Good Morning")
print("How ARe You!!!")
```

### Q. Write a program to take a single digit number from the key board and print is value in English word?

```python
# Get input from the user
user_input = input("Enter a single-digit number: ")

# Check if the input is a single-digit number
if user_input.isdigit() and len(user_input) == 1:
    digit = int(user_input)

    # Use if, elif, else to determine the English word
    if digit == 0:
        print("Zero")
    elif digit == 1:
        print("One")
    elif digit == 2:
        print("Two")
    elif digit == 3:
        print("Three")
    elif digit == 4:
        print("Four")
    elif digit == 5:
        print("Five")
    elif digit == 6:
        print("Six")
    elif digit == 7:
        print("Seven")
    elif digit == 8:
        print("Eight")
    elif digit == 9:
        print("Nine")
else:
    print("Invalid input. Please enter a single-digit number.")
```

## <u> Iterative Statement </u>
If we want to execute a group of statements multiple times then we should go for Iterative statements.<br>

Python supports 2 types of iterative statements.
1. for loop
2. while loop

### For Loop:
If we want to execute some action for every element present in some sequence(it may be string or collection)then we should go for for loop.<br>

Syntax:<br>
for x in sequence :<br>
 body<br>
where sequence can be string or any collection.<br>
Body will be executed for every element present in the sequence<br>



```python
s="Rahul Gandi"
for x in s:
    print(x)
```

### To Print characters Present in string index Wise:

```python
s = input("Enter The Some String")
i = 0
for x in s:
    print("The Character present at",i,"index is:",x)
    i= i +1
```

### While Loop :
if we want to execute a group of statements iteratively until some condition false,then we should go for while loop.

Syntax:<br>
 while condition :<br>
 body
