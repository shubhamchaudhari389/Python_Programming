# <center> Tuple Data Type </center>
1. Tuple is exactly same as List except that it is immutable. i.e once we creates Tuple object,we cannot perform any changes in that object.Hence Tuple is Read Only version of List.<br>
2. If our data is fixed and never changes then we should go for Tuple.<br>
3. Insertion Order is preserved<br>
4. Duplicates are allowed<br>
5. Heterogeneous objects are allowed.<br>
6. We can preserve insertion order and we can differentiate duplicate objects by using index. Hence index will play very important role in Tuple also.Tuple support both +ve and -ve index. +ve index means forward direction(from left to right) and -ve index means backward direction(from right to left)<br>
7. We can represent Tuple elements within Parenthesis and with comma seperator. Parenethesis are optional but recommended to use.


```python
t =10, 20 ,30, 40
print(t)
print(type(t))
```

```python
t=()
print(type(t))
```

## Accessing Elements Of Tuples
We Can Access Either By Index Or By Slice Operator

1. By Using Index:


```python
t =(10, 20, 30 , 40, 50, 60 )
print(t[0])
print(t[-1])
# print(t[100]) # TupeError : Tuple Index Out OF Range 
```

2. By Using Slice Operator

```python
t =(10,20,30,40,50,60)
print(t[2:5])
print(t[2:100])
print(t[::2])
```

### Tuple Packing And Unpacking <br>
We can create a tuples by packing a group of variables.


```python
a =10
b =20
c =30
d =40
t = a,b,c,d
print(t)
```

Tuples UnPacking is The Reverse Process Of Tuples Packing<br>
We Can Unpack A Tuples And ASSIGN its Valuee To Different Variable




```python
t =(10,20,30,40)
a,b,c,d=t
print("a=",a,"b=",b,"c=",c,"d=",d)
```

Note: At the time of tuple unpacking the number of variables and number of values 
should be same. ,otherwise we will get ValueError

### Q. Write a program to take a tuple of numbers from the keyboard and print its sum and  average?


```python
t = eval(input("Enter The Tuples Of Number:"))
l = len(t)
sum = 0
for x in t:
    sum = sum+x
print("The Sum =",sum)
print("The Average =",sum/l)
```

## Differences between List and Tuple:
List and Tuple are exactly same except small difference: List objects are mutable where as Tuple objects are immutable.
In both cases insertion order is preserved, duplicate objects are allowed, heterogenous objects are allowed, index and slicing are supported.
